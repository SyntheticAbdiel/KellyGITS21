```python
import pandas as pd
import geopandas as gp
import numpy as np
import matplotlib.pyplot as plt
```

```python
midterm = pd.read_csv('finalproj/aqimerge.csv')
```

```python
midterm
```

```python
midterm['lon'] = [-122.33, -122.60, -118.23, -104.99, -111.89, -112.07, -87.63, -90.20, -96.80, -71.06, -77.04, -80.19, -122.95, -78.48]
midterm['lat'] = [47.61, 37.77, 34.05, 39.74, 40.76, 33.45, 41.88, 38.63, 32.78, 42.36, 38.91, 25.76, 46.72, 38.03]
```

```python
midterm
```

```python
cities = gp.GeoDataFrame(midterm, geometry = gp.points_from_xy(midterm.lon, midterm.lat))
cities.geometry
```

```python
filepath = 'finalproj/states.shp'
states = gp.read_file(filepath)
```

```python
states.crs
```

```python
states.crs
```

```python
states.head()
```

```python
states = states.to_crs("epsg:4326") #Mercator-projection
```

```python
cities
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax, color="lightgray", edgecolor="grey", linewidth=0.4, legend = True)
cities.plot(ax=ax, color='red', markersize='Poor_Days2020', alpha=1.0, categorical=False, legend = True)
ax.axis()
plt.title('Air Quality Index Poor Days 2020', size=20)
plt.savefig('finalproj/Aqi2020.png')
#plt.axis()
#plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
states.plot(ax=ax, color="lightgray", edgecolor="grey", linewidth=0.4, legend = True)
cities.plot(column= 'Zone', ax=ax, markersize='Poor_Days2010', alpha=1.0, categorical=False, legend = True)
ax.axis()
plt.title('Air Quality Index Poor Days 2010', size=20)
plt.savefig('finalproj/Aqi2010.png')
#plt.axis()
#plt.show()
```

```python
import folium

m = folium.Map(location=[34,-100], titles="OpenStreetMap", zoom_start=4)

m
```

```python
import folium

for i in range(0,len(cities)):
    folium.Circle(
        location=[cities.iloc[i]['lat'], cities.iloc[i]['lon']],
        popup=cities.iloc[i]['City'],
        radius=float(cities.iloc[i]['Poor_Days2020'])*2000,
        color='crimson',
        fill=True,
        fill_color='crimson',
    ).add_to(m)

m
```

```python
m.save('finalproj/aqidata2020.html')
```
